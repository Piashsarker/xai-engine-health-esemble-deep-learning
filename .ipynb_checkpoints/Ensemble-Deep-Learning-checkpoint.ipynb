```python
import pandas as pd
import numpy as np

# ML
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# Models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# Explainability
import shap
from lime.lime_tabular import LimeTabularExplainer

import matplotlib.pyplot as plt
```

```python
# -----------------------------
# 1. LOAD DATA
# -----------------------------
df = pd.read_csv("dataset/augmented_data_with_environment.csv")

print("Dataset shape:", df.shape)
print(df.head())
```

```python
# -----------------------------
# 2. PREPROCESSING
# -----------------------------
# Assume last column is target
target_col = df.columns[-4]

X = df.drop(columns=[target_col])
y = df[target_col]
```

```python
# Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)
```

```python

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# -----------------------------
# 3. ENSEMBLE MODEL (STACKED MODEL 1)
# -----------------------------
base_models = [
    ('rf', RandomForestClassifier(n_estimators=100)),
    ('svm', SVC(probability=True)),
    ('gb', GradientBoostingClassifier()),
    ('dt', DecisionTreeClassifier()),
    ('knn', KNeighborsClassifier())
]

meta_model = LogisticRegression()

stack_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    passthrough=False
)
```

```python

# Train
stack_model.fit(X_train_scaled, y_train)
```

```python
# -----------------------------
# 4. EVALUATION
# -----------------------------
y_pred = stack_model.predict(X_test_scaled)
y_prob = stack_model.predict_proba(X_test_scaled)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# AUC (multi-class)
auc = roc_auc_score(y_test, y_prob, multi_class='ovr')
print("\nAUC:", auc)
```

```python
# -----------------------------
# 5. LIME EXPLANATION
# -----------------------------
explainer_lime = LimeTabularExplainer(
    training_data=X_train_scaled,
    feature_names=X.columns.tolist(),
    class_names=le.classes_.astype(str),
    mode='classification'
)
```

```python


# Explain one instance
i = 0  # change index if needed
exp = explainer_lime.explain_instance(
    X_test_scaled[i],
    stack_model.predict_proba,
    num_features=5
)

print("\nLIME Explanation:")
print(exp.as_list())
```

```python

# Plot LIME
exp.as_pyplot_figure()
plt.title("LIME Explanation")
plt.show()
```

```python
# -----------------------------
# 6. SHAP EXPLANATION
# -----------------------------
# Use KernelExplainer for stacked model
explainer_shap = shap.KernelExplainer(
    stack_model.predict_proba,
    shap.sample(X_train_scaled, 100)
)
```

```python

# Compute SHAP values (small sample for speed)
shap_values = explainer_shap.shap_values(X_test_scaled[:50])
```

```python
# -----------------------------
# 6.1 Global Explanation
# -----------------------------
shap.summary_plot(shap_values, X_test_scaled[:50], feature_names=X.columns)
```

```python
# -----------------------------
# 6.2 Local Explanation
# -----------------------------
shap.force_plot(
    explainer_shap.expected_value[0],
    shap_values[0][0],
    X_test_scaled[0],
    feature_names=X.columns,
    matplotlib=True
)
```

```python

# -----------------------------
# 7. FEATURE IMPORTANCE (SHAP)
# -----------------------------
shap.summary_plot(shap_values, X_test_scaled[:50], plot_type="bar")
```
